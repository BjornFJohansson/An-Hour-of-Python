```python
f,r = parse_primers("""

>f
GATGAGTTCGTGTCCGTACAACT

>r
GGTTATCGAAATCAGCCACAGCG

""")
```

```python
Lambda = genbank("J02459.1")
```

```python
Lambda
```

```python
pcr_prod = pcr(f,r,Lambda)
```
